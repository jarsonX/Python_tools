```python
import pandas as pd
```

```python
# Whatever the source is (csv, excel, db), we need to put it into pandas dataframe

# csv example
input_data = pd.read_csv('table.csv')
input_data
```

```python
# Result in separate columns
# This step is required for the final result table

# errs
# .ne and .shift = check if the next row is equal to current row
# .cumsum = cumulative sum over axis

# df_sep_cols
# simple aggregation and sorting

errs = input_data['error'].ne(input_data['error'].shift()).cumsum()

df_sep_cols = (input_data.groupby([errs, 'error'], as_index=False)
                .agg(MIN=('seq num','min'), MAX=('seq num','max'))
                .sort_values('error', ignore_index=True))
df_sep_cols
```

```python
#Result showing ranges

df_ranges = (df_sep_cols['MIN'].astype(str).str.cat(df_sep_cols['MAX'].astype(str), sep='-')
             .groupby(df_sep_cols['error']).agg('; '.join)
             .reset_index(name='range'))
df_ranges
```
